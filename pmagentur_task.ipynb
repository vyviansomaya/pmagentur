```python
# 1.1 Factorial

def factorial(num):
    sum=1
    for i in range(1,num+1):
        sum*=i
    return sum
```

```python
print(factorial(8))
```

```python
#1.3 prime factorial 

import math

def ifprime(num):
    flg=True
    for i in range(2,math.ceil(num/2)):
        if num%i==0:
            flg=False
            break
    return flg
            
        
    
        
```

```python
def getPrimeNumbersTo(num):
    
    lst_prime=[]
    if(num==2):
        lst_prime.append(num)
    elif num>2:
        for i in range (2,num):
            if(ifprime(i)):
                lst_prime.append(i)
    str1="";
    str1=','.join(map(str,lst_prime))
    return str1
```

```python
getPrimeNumbersTo(13)
```

```python
#1.4 binary tree

import re
def constructTree(strArr):
    node={}
    child={}
    for i in strArr:
        lst2=i.split(",")
        
            
        a=re.sub(r"[()]","",lst2[0])
        b=re.sub(r"[()]","",lst2[1])
      #  print("a", a)
      #  print("b ",b)
        if a in child.keys():
      #      print("dis1")
            return False
        else:
            child[a]=b
       #     print(child)
        if b not in node.keys():
            node[b]=[a]
        #    print(node)
        else:
            if(len(node.get(b))>=2):
         #       print("dis2")
                return False
            else:
                node.get(b).append(a)
          #      print(node)
    return True
        
```

```python
lst1=["(1,2)","(2,4)","(5,7)","(7,2)","(9,5)"]
constructTree(lst1)
```
